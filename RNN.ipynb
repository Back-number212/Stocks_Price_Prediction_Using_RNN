```python
scaler = MinMaxScaler(feature_range=(0, 1))
bnb_scaled = scaler.fit_transform(bnb.values.reshape(-1, 1))

def create_dataset(dataset, time_step=365):
    X, y = [], []
    for i in range(len(dataset) - time_step):
        X.append(dataset[i:i + time_step, 0])
        y.append(dataset[i + time_step, 0])
    return np.array(X), np.array(y)

time_step = 365
train_size = int(len(bnb_scaled) * 0.8)
X_train, y_train = create_dataset(bnb_scaled[:train_size], time_step)
X_test, y_test = create_dataset(bnb_scaled[train_size - time_step:], time_step)
X_train = X_train.reshape(X_train.shape[0], X_train.shape[1], 1)
X_test = X_test.reshape(X_test.shape[0], X_test.shape[1], 1)

model = Sequential()
model.add(SimpleRNN(50, return_sequences=True, input_shape=(time_step, 1)))
model.add(SimpleRNN(50))
model.add(Dense(1))
model.compile(optimizer='adam', loss='mean_squared_error')
model.fit(X_train, y_train, epochs=10, batch_size=32, verbose=1)

train_pred = scaler.inverse_transform(model.predict(X_train))
test_pred = scaler.inverse_transform(model.predict(X_test))
y_train_actual = scaler.inverse_transform(y_train.reshape(-1, 1))
y_test_actual = scaler.inverse_transform(y_test.reshape(-1, 1))
train_rmse = np.sqrt(mean_squared_error(y_train_actual, train_pred))
test_rmse = np.sqrt(mean_squared_error(y_test_actual, test_pred))
print(f'Train RMSE: {train_rmse}')
print(f'Test RMSE: {test_rmse}')
future_days = 366
future_predictions = []
last_days = bnb_scaled[-time_step:]
for _ in range(future_days):
    pred = model.predict(last_days.reshape(1, -1, 1))[0, 0]
    future_predictions.append(pred)
    last_days = np.append(last_days[1:], pred)

future_predictions = scaler.inverse_transform(np.array(future_predictions).reshape(-1, 1))
future_dates = [datetime(2024, 1, 1) + timedelta(days=i) for i in range(future_days)]
comparison_df = pd.DataFrame({
    'Date': bnb.index[-len(test_pred):],
    'Actual': y_test_actual.flatten(),
    'Predicted': test_pred.flatten()})
print(comparison_df.head(10))
bnb1= yf.download('BNB-USD', start='2024-01-01', end='2024-12-31')['Close']
plt.figure(figsize=(14, 7))
plt.plot(bnb1.index, bnb1, label='Actual', color='blue')
plt.plot(bnb.index[-len(test_pred):], test_pred, label='Predict', color='red')
plt.title('Binance SimpleRNN Prediction 2024')
plt.xlabel('Date')
plt.ylabel('Stock Price')
plt.legend()
plt.show()
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.preprocessing import MinMaxScaler
from sklearn.metrics import mean_squared_error
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import LSTM, Dense, Dropout
import yfinance as yf
from datetime import datetime, timedelta

# Tải dữ liệu Tesla từ Yahoo Finance
data = yf.download('TSLA', start='2019-01-01', end='2024-12-31')['Close']
scaler = MinMaxScaler(feature_range=(0, 1))
data_scaled = scaler.fit_transform(data.values.reshape(-1, 1))

def create_dataset(dataset, time_step=60):
    X, y = [], []
    for i in range(len(dataset) - time_step):
        X.append(dataset[i:i + time_step, 0])
        y.append(dataset[i + time_step, 0])
    return np.array(X), np.array(y)

# Cài đặt tham số
time_step = 365
train_size = int(len(data_scaled) * 0.8)
X_train, y_train = create_dataset(data_scaled[:train_size], time_step)
X_test, y_test = create_dataset(data_scaled[train_size - time_step:], time_step)
X_train = X_train.reshape(X_train.shape[0], X_train.shape[1], 1)
X_test = X_test.reshape(X_test.shape[0], X_test.shape[1], 1)

# Xây dựng mô hình LSTM
model = Sequential()
model.add(LSTM(50, return_sequences=True, input_shape=(time_step, 1)))
model.add(Dropout(0.2))
model.add(LSTM(50, return_sequences=False))
model.add(Dropout(0.2))
model.add(Dense(1))
model.compile(optimizer='adam', loss='mean_squared_error')

# Huấn luyện mô hình
model.fit(X_train, y_train, epochs=20, batch_size=32, verbose=1)

# Dự đoán và đánh giá
train_pred = scaler.inverse_transform(model.predict(X_train))
test_pred = scaler.inverse_transform(model.predict(X_test))
y_train_actual = scaler.inverse_transform(y_train.reshape(-1, 1))
y_test_actual = scaler.inverse_transform(y_test.reshape(-1, 1))

# Tính toán RMSE
train_rmse = np.sqrt(mean_squared_error(y_train_actual, train_pred))
test_rmse = np.sqrt(mean_squared_error(y_test_actual, test_pred))
print(f'Train RMSE: {train_rmse}')
print(f'Test RMSE: {test_rmse}')

# Dự đoán tương lai
future_days = 365
future_predictions = []
last_60_days = data_scaled[-time_step:]
for _ in range(future_days):
    pred = model.predict(last_60_days.reshape(1, -1, 1))[0, 0]
    future_predictions.append(pred)
    last_60_days = np.append(last_60_days[1:], pred)

future_predictions = scaler.inverse_transform(np.array(future_predictions).reshape(-1, 1))
future_dates = [datetime(2024, 1, 1) + timedelta(days=i) for i in range(future_days)]

# Tạo DataFrame so sánh
comparison_df = pd.DataFrame({
    'Date': data.index[-len(test_pred):],
    'Actual': y_test_actual.flatten(),
    'Predicted': test_pred.flatten()})
print(comparison_df.head(10))

# Hiển thị kết quả dự đoán và so sánh
plt.figure(figsize=(14, 7))
data1 = yf.download('TSLA', start='2024-01-01', end='2024-12-31')['Close']
plt.plot(data1.index, data1, label='Actual (BNB)', color='blue')
plt.plot(data.index[-len(test_pred):], test_pred, label='Predict (Tesla)', color='red')
plt.title('Tesla LSTM Prediction 2024')
plt.xlabel('Date')
plt.ylabel('Price')
plt.legend()
plt.show()
```
